# Deploy Face dectection with Gradio


```python
from deepface import DeepFace
import gradio as gr

def verify(img1, img2):
    result = DeepFace.verify(img1_path = img1, img2_path = img2)
    return result

def detect(img):
    dct = DeepFace.detectFace(img)
    return dct

demo = gr.Blocks()

with demo:
    with gr.Tab("Image input"):
      with gr.Row():
          image_input = gr.Image()
          image_output = gr.Image()
      image_button = gr.Button("Detect")
    with gr.Tab("Webcam input"):
      webcam_input = gr.Image(source="webcam")
      webcam_output = gr.Image()
      webcam_button = gr.Button("Detect")

    image_button.click(detect, inputs=image_input, outputs=image_output)
    webcam_button.click(detect, inputs=webcam_input, outputs=webcam_output)

    imgToVerify = gr.Image()
    verifyOutput = gr.Textbox()

    verifyImgButton = gr.Button("Verify with Image input")
    verifyCamButton = gr.Button("Verify with Webcam input")

    verifyImgButton.click(verify, inputs=[image_output,imgToVerify], outputs = verifyOutput)
    verifyCamButton.click(verify, inputs=[webcam_output,imgToVerify], outputs = verifyOutput)

demo.launch(share=True)
```
